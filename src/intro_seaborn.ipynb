# Introduction to Seaborn

This a high level python module that allows an easy way of managing our data and visualize it without many code lines.

## Imports

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

## First plotting steps

````python
print(*sns.get_dataset_names())
penguins = sns.load_dataset('penguins'); print(penguins.columns)
````

````python
# Plotting the relation x=bill_length_mm, y=bill_depth_mm
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    # Change the color of all data points
    #color='purple',
    # This parameter allows the assignment of each dataset to a color
    hue='species'
)
plt.tight_layout()
plt.show()
````

| Atributo visual	| Descripción	| Variable (asociada a columna) |	Valor constante	| Notas |
| - | - | - | - | - |
| Color |	Color de los puntos |	hue='columna' |	color='red' / hex / RGB |	hue genera leyenda automática; color fija todos los puntos. |
| Tamaño	| Tamaño de los puntos |	size='columna' |	s=50 o size=30	| size genera leyenda automática; s es solo un valor fijo. |
| Forma / marcador |	Forma de los puntos |	style='columna' |	marker='o' |	style solo para categórico; marker fija la forma. |
| Transparencia |	Nivel de opacidad de los puntos |	solo manual (Matplotlib) |	alpha=0.5	| No genera leyenda automática por variable. |
| Borde / edgecolor / linecolor |	Color del borde de los puntos |	solo manual |	edgecolor='black'	| Solo define el borde; no genera leyenda automática. |

````python
# Combining attributes
sns.scatterplot(data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",           # color according to the species
    size="flipper_length_mm", # Size of the bill length
    style="sex",             # Shape according to the sex
    edgecolor="purple",        # Constant border
    alpha=0.5                   # Transparency
)
plt.tight_layout()
plt.show()
````

### Doing a histogram

````python
sns.histplot(
    data=penguins,
    x="body_mass_g",
    edgecolor="white", # Border color
    alpha=0.3, # Transparency
    bins=20 # Number of barras
)
plt.tight_layout()
plt.show()
````

### Let's see how a boxplot works

````python
sns.boxplot(
    data=penguins,
    x= "sex",
    y="body_mass_g"
)
plt.tight_layout(); plt.show()

# Cambiando el color de las cajas y el de las líneas
sns.boxplot(
    data=penguins,
    x= "sex",
    y="body_mass_g",
    color="forestgreen",
    boxprops=dict(edgecolor="black", alpha=0.5) # Select an edge color
)
plt.tight_layout(); plt.show()
````

## How to add titles or labels?

For doing so, it is easier (and sometimes needed) to use `matplotlib.pyplot` to add legends, titles or labels.

To see the documentation of a command we shall use:

```python
help(<function>)

# Other option
?<function>
```

````python
sns.scatterplot(data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",           # Color according to species
    size="flipper_length_mm", # Size according to flipper length
    style="sex",             # Shape according to sex
    edgecolor="purple"        # Constante edge color
)

# Agregar títulos y etiquetas
plt.title(
    "Relation between bill length and depth", 
    fontsize=14, 
    fontweight="bold",
    fontstyle = "oblique", 
    color="black"
)
plt.xlabel("Bill length (mm)", fontsize=12, color='black')
plt.ylabel("Bill depth (mm)", fontsize=12, color='black')

# Move legend as desired
plt.legend(
    bbox_to_anchor=(1.02, 1),   # Position outside the axe
    loc='upper left',           # Legend location/position
    borderaxespad=0,
    fontsize = 8,
    title_fontsize=10,
    labelcolor = "black"
)

# Modify plot background and figure axs
plt.gca().set_facecolor("mintcream")    # Axs background
plt.gcf().set_facecolor("mintcream")   # Figure background

plt.grid(True, linestyle=":", alpha=0.9)

plt.tight_layout()
plt.show()
````

## For doing more than one plot

To ease this kind of tasks we should use a dictionary with our style preferences for plotting, so that we only call to that key and access its corresponding value.

````python
letter_style = {
    'fontsize': 14,
    'color': 'darkblue',
    'fontweight': 'bold',
    'fontstyle': 'italic'
}

sns.scatterplot(data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",           # Color according to species
    size="flipper_length_mm", # Size according to flipper length
    style="sex",             # Shape according to sex
    edgecolor="purple"        # Constante edge color
)

plt.title("Relation between bill length and depth", fontdict=letter_style)

# Move legend as desired
plt.legend(
    bbox_to_anchor=(1.02, 1),   # Position outside the axe
    loc='upper left',           # Legend location/position
    borderaxespad=0,
    fontsize = 8,
    title_fontsize=10,
    labelcolor = "black"
)

# Modify plot background and figure axs
plt.gca().set_facecolor("mintcream")    # Axs background
plt.gcf().set_facecolor("mintcream")   # Figure background

plt.grid(True, linestyle=":", alpha=0.9)

plt.tight_layout()
plt.show()
````

## Another way to access the fig and axs objects

````python
letter_style = {
    'fontsize': 14,
    'color': 'darkblue',
    'fontweight': 'bold',
    'fontstyle': 'italic'
}

# Asign the ax object to a variable 
g = sns.scatterplot(data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",           # Color according to species
    size="flipper_length_mm", # Size according to flipper length
    style="sex",             # Shape according to sex
    edgecolor="purple"        # Constante edge color
)

g.set_title(
    'Relation between bill length and depth', 
    fontsize=14, 
    fontweight="bold", 
    fontstyle="italic", 
    color="darkblue"
)

g.set_xlabel(
    "Bill length (mm)", 
    fontsize=12
)
g.set_ylabel(
    "Bill depth (mm)", 
    fontsize=12
)

g.figure.suptitle("Visualizing morphological relations among penguins", fontsize=16, fontweight="bold")

plt.tight_layout()
plt.show()
````
