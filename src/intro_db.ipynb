# Introduction to related data bases

A data base is a colection of data sets related by some related characteristics that links them.

## Database terminology

1. Row or tuple. Unique identifier called *primary key*.
2. Column or attribute. The *primary key* itself.
3. *Foregin key*. Pointer or link to a *primary key* from an outsider table.
4. *Field*. Intersection between a row and a column. ***Admits multiple relations among tables.***
5. Optional *field*. It can be empty.
